__Nested Dictionaries__

```python
myfamily={"chidima":{"name": "Emil", "Year": 2004},
          "Tabitha":{"name":"Bob", "Year": 2020},
          "Christiana":{"name":"Mercy","Year": 2005}}
myfamily
```

```python
print(myfamily["chidima"],["name"])
```

```python
for x, obj in  myfamily.items():
    print(x)
    for y in obj:
        print(y +':', obj[y])
```

__Dictionary Methods__

| Method|Discription                                               |
|-------------|---------------------------------------------------------|
|`fromkeys()`    |Return a dictionary with the specified keys and values. </br>  |
|`clear()`        | Remove all the element from the list. </br> |
|`copy()`  |Return a copy of the list . </br> |
|`items()` | Return the list containing a tuple for each key value pair. </br> |
|`keys()`  |Return the list containing dictionary's keys.</br> |
|`pop()` | Remove the element at the specified position. </br |
|`popitems()` |  Remove the last inserted key-value pair . </br>|
|`setdefault()`    |Return the value of the specified key. </br>|
|`update()`     |Update the dictionary with the specified key-value pairs. </br>|
|`values()`     |Return list of all the values in the dictinary. </br>|
