Polymorphism
 ======================

Polymorphism is method /functions/Operators with the same name that can be execute on many object or classes.

#### Class Polymorphism
polymorphism is often used in Class methods, wher we can have multiple classes with the same method name

> Employee Management App

```python
class Employee:
  no_employees = 0
  def __init__(self, name: str, age: int, salary: float):
    self.name = name
    self.age = age
    self.salary = salary
    Employee.no_employees += 1
    self.id = Employee.no_employees

  def __str__(self):
    return f"Id: {self.id}\nName: {self.name}\nAge: {self.age}\nSalary: {self.salary}"
```

```python
class EmployeesManager:
  def __init__(self):
    self.employees = []

  def add_employee(self, name, age, salary):
    employee = Employee(name, age, salary)
    self.employees.append(employee)

  def remove_employee(self, id: int):
    for employee in self.employees:
        if id == employee.id:
          self.employees.remove(employee)
          break
    else:
      print("Employee Not Found!")

  def update_employee(self, id, age, salary):
    for employee in self.employees:
        if id == employee.id:
          employee.age = age
          employee.salary = salary
          break
    else:
      print("Employee Not Found!")

  def list_employess(self):
    for employee in self.employees:
      print(employee)
      print("-" * 50)
```

Encapsulation
------------------

Encapsulation is about protecting data inside a class.

- **Data Protection**: Prevent accidental modification of data
- **Validation**: validate data before setting it
- **Flexibility**: Internsl implementation can change without affecting external code
- **Control**: full control over how data can is accessed and modified.

- **Private Properties** : used a double underscore _ _ prefix
  > `NB`   private properties can not be accessed directly from outside the class.

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.__age=age  #private property

p1=Person("Emil", 25)
print(p1.name)
print(p1.__age)        #thsis willl cause error
```

```python
class Person:
    def __init__(self,name,age):
        self.name=name
        self.__age=age  #private property
    def get_age(self):       #using get method
        return self.__age
    

p1=Person("Emil", 25)
print(p1.get_age())
      
```

```python
class Student:
    def __init__(self,name):
        self.name=name
        self.__grade= 0
    def set_grade(self,grade):
        if 0 <= grade <= 100:
            self.__grade=grade
        else:
            print("Grade must be beteen 0 and 100")

    def get_grade(self):
        return self.__grade
    
    def get_status(self):
        if self.__grade >= 60:
            return "Passed"
        else:
            return "Failed"
student=Student("Emil")
student.set_grade(85)
print(student.get_grade())
print(student.get_status())
```

Python Inner Classes
-----------------

An inner class is class define inside another class.The Inner clsas can access the properties and method of the outer class.

```python
class Outer:
    def __init__(self):
        self.name="Outer class"

    class Inner:
        def __init__(self):
            self.name="Inner class"

        def display(self):
            print("This is the Inner class")

outer=Outer()
print(outer.name)
```

```python
class Outer:
    def __init__(self):
        self.name="Outer class"

    class Inner:
        def __init__(self,outer):
            self.outer=outer

        def display(self):
            print("This is the Inner class")

outer=Outer()
inner=outer.Inner(outer)                      
inner.display()
```
