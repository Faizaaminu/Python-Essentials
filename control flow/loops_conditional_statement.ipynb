#### Elif
the elif keyword says "if previous conditions were not true, then try this condition. it allow you to check multiple expressions for `True` and execute a block of code as soon as one of the condition is evaluate to  `true`

```python
if 25 < 10:
    print('it work!')
elif 25 < 30:
    print('worked')
else:
    print('it did not work!')
```

__Logical Operators__

logical operators are used to combine conditional statement.
- `and` Return True if both statement are true
- `or` Return True if one of the statement is true
- `not` Reverses the result,returns False if the result is true

```python
""" Test if a is greater than b, AND c is greater than a"""
a=200
b=33
c=500
if a >b and c > a:
    print("Both condition are True")
```

```python
""" Test if a is greater than b, OR a is greater than c"""
a=200
b=33
c=500
if a >b or a > c:
    print("At least one of the condition is True")
```

```python
""" Test if a is NOT greater than b """
a=33
b=200

if not a > b :
    print("a is not greater than b")
```

__The Match Statement__

_the match statement select one of many code block to be executed_
- the `match` expression is evaluated once
- the value of the expression is compared with the values of each case

```python
day=4
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesdat")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
```

```python
day=4
match day:
    case 6:
        print(" Today is Saturday")
    case 7:
        print("Today  is Sunday")
    case _:                       ##Use underscore _ as the last case value to execute code where ther is not other matches
        print("Looking forward to the weekend")
```

### While Loops


iterate over a sequence if condition  
`while` loop execute a set of statement as long as a condition is true

```python
number=1   #variable
while number < 5:
    print(number)
    number = number + 1
```

```python
number=1
while number < 5:
    print(number)
    if number  == 3:
        break
    number = number + 1
```

```python
number=1
while number < 5:
    print(number)
    if number  == 6:
        break   #break discontinue
    number = number + 1
else:
    print('No longer < 5')
```

```python
number=1
while number < 5:
    number = number + 1
    if number  == 3:
      continue    #skip
    print(number)
else:
    print('No longer < 5')
```

`NB`  loop control Statement.
  >loop control statement change execution from their normal sequence
  - __continue__ Statement: it returns the control to the begining of the loop.
  -  __Break__ Statement: it brings control out of the loop.
  -  __Pass__ to write empty loop
