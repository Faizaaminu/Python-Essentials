The **Walrus** Operator: `:=` assigns values variables as part of a larger expression

```python
numbers=[1,2,3,4,5]
count=len(numbers)
if count >3:
    print(f"List has {count} element")

    if(count:=len(numbers)) > 3:
        print(f"List has {count} element")
```

Identity Operators
------------------------

Identity operators are used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location.

| Operators|Description                                              |
|-------------|---------------------------------------------------------|
|`is`    |Return True if both variables are the same object. </br>  |
|`is not` |Return True if both variables are not the same object . </br> |




```python
x=["apple", "banana"]
y=["apple", "banana"]
z=x

print(x is z)
print( x is y)
print( x == y)
print(x is not y)
```

Bitwise Operators
---------------------------

Bitwise operators are used to compare (binary) numbers

| Operators|Description|Description                                           |     
|-------------|-----------------------------------------|----------------|
|`&`    |AND. </br>  |Set each bit to 1 if both bits are 1. </br>
|`!`      |OR. </br>  |Set each bit to 1 if one of two  bits are 1. </br>
|`^`    |XOR </br>  |Set each bit to 1 if only one of two bits are 1. </br>
|`~`    |NOT. </br>  |Invert all the bits. </br>
|`<<`    |Zero fill left shift. </br>  |Shift left by pushing zeroes in from the right and let the leftmost bits fall off. </br>
|`>>`    |Signed right shift. </br>  |Shift right by pushing copies of the leftmost bit in from the left, and let the rightmost bits fall off. </br>





```python
#The & operator compare each bit to 1 if both are 1 , otherwise it is set to zero
print(6 & 3)
```

`NB` The binary representation of 6 is 0110 and the binary representation of 3 is 0011.
> The & operator compare the bit and return 0010, which is 2 decimal.

```python
#The  | operator compare each bit to 1 if one or both are 1 , otherwise it is set to zero

print(6 |3 )
```

`NB` The binary representation of 6 is 0110 and the binary representation of 3 is 0011.
> The & operator compare the bit and return 0111, which is 7 decimal.

```python
#The  ^ operator compare each bit to 1 if only one is 1 , otherwise (if both are 1 or  both are 0) it is set to 0

print(6 ^ 3 )
```

`NB` The binary representation of 6 is 0110 and the binary representation of 3 is 0011.
> The & operator compare the bit and return 0101, which is 5 decimal.
