`strftime()` method

```python
"""" Display the name of the month"""

x=datetime.datetime(2024,5,17)
print(x.strftime("%B"))
```

| Directice |Discription                                               |
|-------------|---------------------------------------------------------|
|`%a`    |Weekday. </br> short version |
|`%A`        | Weekday. </br>full version.  </br> |
|`%w`  |Weekday . </br>As a number 0-6,0 is sunday. </br>. |
|`%d` | Day of month. </br>01-31. </br> |
|`%b`  | Month name. </br>fsjort version. </br> |
|`%B` | Month name. </br>full name. </br> |
|`%m`   |Month . </br>As a number 01-12</br>|
|`%y`    |Year. </br>short version, without century. </br>|
|`%Y`     |Year. </br>full version. </br>|
|`%H`     |Hour. </br>00-23. </br>|
|`%i`     |Hour. </br>00-12. </br>|
|`%p`     |AM/PM.  </br>|
|`%M`     |Minute. </br>00-59 </br>|
|`%S`     |Second. </br>00-59 </br>|
|`%f`     |Microsecond. </br>000000-999999</br>|
|`%z`    |UTC offset. </br>|
|`%Z`     |Timezone. </br>00-59 </br>|
|`%j`    |Day . </br>number of year 001-366</br>|
|`%U`  |Week br>number of year, Sunday as the first day of week, 00-53 </br>|
|`%W`   |Week/br>number of year, Monday as the first day of week, 00-53</br>|
|`%c` |Local version of date and time</br>|
|`%C`  |Century.</br>|
|`%x`   |Local version of time</br>|
|`%X`   |Local version of time </br>|
|`%%`   |A % character.</br>|
|`%G`  |ISO 8601 YEAR </br>|
|`%u`   |ISO 8601 weekday(-7)</br>|
|`%V`     |ISO 8601 weeknumber(01-53)</br>|

```python
"""" Display the name of theshort version of weekday"""

x=datetime.datetime(2024,5,17)
print(x.strftime("%a"))
```

```python
"""" Display the name of thefull   version  of weekday"""

x=datetime.datetime.now()
print(x.strftime("%A"))
```

```python
"""" Display the Week  number   of a  year"""

x=datetime.datetime.now()
print(x.strftime("%U"))
```

```python
"""" Display the name of thefull   version  of weekday"""

x=datetime.datetime.now()
print(x.strftime("%A,%d,%B,%Y %H:%M:%S"))
```
