__Built-in Math Functions__

>The`min()` and  `max()` functions can be used to find the lowest or highest value in an iterable.

```python
x=min(5, 10,25)
y=max(5,10,25)
print(x)
print(y)
```

> The `abs()` function returns the absolute(positive) value of the specified number.

```python
x=abs(-7.25)
x
```

> The `pow(x,y)` function returns the value of x to the power of y (x^y).

```python
""" Return the value of 4 to the power of 3"""

x=pow(4,3)
x
```

```python
import math    ## math module
```

>The `math.sqrt()` method return the square root of a number.

```python
## math,sqrt return the square root of a number

x=math.sqrt(64)
x
```

>The `math.ceil()` method round a number upward to its nearest integer. </br>
>The `math.floor()` method round a number downwards to its nearest integer and return the result.

```python
x=math.ceil(1.4)
y=math.floor(1.4)
print(x)
print(y)
```

> The `math.pi` constant, returns the value of pi(3.14).

```python
x=math.pi
x
```

```python
# Math.log() takes the natural logarithms of its argument

math.log(2.7182)
```

```python
#Add a second argument to specify the base

math.log(100,10)  #Take the log base 10 of 100
```

```python
# math.exp() raises e to the power of its element
math.exp(10)
```

>`round()` function let you round numbers to the nearest whole number

```python
#Use round() to round a number to the nearest whole number

round(233.234)
```

```python
# Add a second argument to round to a specified decimal place

round(233.234, 1)  #round to 1 decimal place
```

```python
#Enter a negative number to round to the left of the decimal

round(233.234,-1)  #round to the 10,s place
```
