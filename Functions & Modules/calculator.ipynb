```python
class StudentRecordSystem:
    def __init__(self):
        self.students = []
        self.next_id = 1
    
    def calculate_grade(self, scores):
        """Calculate grade based on average score"""
        if not scores:
            return 'F'
        
        average = sum(scores) / len(scores)
        
        if average >= 90:
            return 'A'
        elif average >= 80:
            return 'B'
        elif average >= 70:
            return 'C'
        elif average >= 60:
            return 'D'
        else:
            return 'F'
    
    def add_student(self):
        """Add a new student record"""
        print("\n--- Add New Student ---")
        
        # Collect basic information
        name = input("Enter student name: ").strip()
        
        # Validate age
        while True:
            try:
                age = int(input("Enter student age: "))
                if age < 5 or age > 100:
                    print("Please enter a valid age (5-100)")
                    continue
                break
            except ValueError:
                print("Please enter a valid number for age")
        
        course = input("Enter course: ").strip()
        
        # Collect scores
        scores = []
        print("\nEnter subject scores (enter -1 to finish):")
        while True:
            try:
                score = float(input("Enter score (0-100): "))
                if score == -1:
                    break
                if score < 0 or score > 100:
                    print("Please enter a score between 0 and 100")
                    continue
                scores.append(score)
            except ValueError:
                print("Please enter a valid number")
        
        # Calculate grade
        grade = self.calculate_grade(scores)
        
        # Create student record
        student = {
            'id': self.next_id,
            'name': name,
            'age': age,
            'course': course,
            'scores': scores,
            'grade': grade
        }
        
        self.students.append(student)
        self.next_id += 1
        
        print(f"\nStudent added successfully! Student ID: {student['id']}")
        print(f"Name: {name}, Course: {course}, Grade: {grade}")
    
    def view_all_students(self):
        """Display all student records"""
        print("\n--- All Student Records ---")
        
        if not self.students:
            print("No student records found.")
            return
        
        print(f"{'ID':<5} {'Name':<20} {'Age':<5} {'Course':<15} {'Grade':<5}")
        print("-" * 60)
        
        for student in self.students:
            print(f"{student['id']:<5} {student['name']:<20} {student['age']:<5} "
                  f"{student['course']:<15} {student['grade']:<5}")
    
    def search_student(self):
        """Search for a student by ID or name"""
        print("\n--- Search Student ---")
        
        if not self.students:
            print("No student records found.")
            return
        
        search_term = input("Enter student ID or name to search: ").strip()
        
        # Try to search by ID first
        try:
            search_id = int(search_term)
            for student in self.students:
                if student['id'] == search_id:
                    self._display_student_details(student)
                    return
        except ValueError:
            pass  # Not an ID, search by name
        
        # Search by name
        found_students = []
        for student in self.students:
            if search_term.lower() in student['name'].lower():
                found_students.append(student)
        
        if found_students:
            print(f"\nFound {len(found_students)} matching student(s):")
            for student in found_students:
                self._display_student_details(student)
        else:
            print("Student not found.")
    
    def _display_student_details(self, student):
        """Display detailed information about a student"""
        print(f"\nStudent ID: {student['id']}")
        print(f"Name: {student['name']}")
        print(f"Age: {student['age']}")
        print(f"Course: {student['course']}")
        if student['scores']:
            print(f"Scores: {', '.join(map(str, student['scores']))}")
            print(f"Average Score: {sum(student['scores'])/len(student['scores']):.2f}")
        else:
            print("Scores: No scores available")
        print(f"Grade: {student['grade']}")
        print("-" * 40)
    
    def edit_student(self):
        """Edit an existing student record"""
        print("\n--- Edit Student Record ---")
        
        if not self.students:
            print("No student records found.")
            return
        
        try:
            student_id = int(input("Enter student ID to edit: "))
        except ValueError:
            print("Please enter a valid student ID (number)")
            return
        
        # Find student
        student = None
        for s in self.students:
            if s['id'] == student_id:
                student = s
                break
        
        if not student:
            print("Student not found.")
            return
        
        self._display_student_details(student)
        
        # Edit fields
        print("\nLeave field blank to keep current value:")
        
        new_name = input(f"New name [{student['name']}]: ").strip()
        if new_name:
            student['name'] = new_name
        
        new_age = input(f"New age [{student['age']}]: ").strip()
        if new_age:
            try:
                student['age'] = int(new_age)
            except ValueError:
                print("Invalid age, keeping current value")
        
        new_course = input(f"New course [{student['course']}]: ").strip()
        if new_course:
            student['course'] = new_course
        
        # Option to update scores
        update_scores = input("\nUpdate scores? (y/n): ").strip().lower()
        if update_scores == 'y':
            new_scores = []
            print("Enter new subject scores (enter -1 to finish):")
            while True:
                try:
                    score = float(input("Enter score (0-100): "))
                    if score == -1:
                        break
                    if score < 0 or score > 100:
                        print("Please enter a score between 0 and 100")
                        continue
                    new_scores.append(score)
                except ValueError:
                    print("Please enter a valid number")
            
            if new_scores:
                student['scores'] = new_scores
                student['grade'] = self.calculate_grade(new_scores)
        
        print("\nStudent record updated successfully!")
        self._display_student_details(student)
    
    def delete_student(self):
        """Delete a student record"""
        print("\n--- Delete Student Record ---")
        
        if not self.students:
            print("No student records found.")
            return
        
        try:
            student_id = int(input("Enter student ID to delete: "))
        except ValueError:
            print("Please enter a valid student ID (number)")
            return
        
        # Find and delete student
        for i, student in enumerate(self.students):
            if student['id'] == student_id:
                self._display_student_details(student)
                confirm = input("\nAre you sure you want to delete this student? (y/n): ").strip().lower()
                
                if confirm == 'y':
                    deleted_student = self.students.pop(i)
                    print(f"\nStudent '{deleted_student['name']}' (ID: {deleted_student['id']}) deleted successfully!")
                    return
                else:
                    print("Deletion cancelled.")
                    return
        
        print("Student not found.")
    
    def calculate_grade_standalone(self):
        """Standalone grade calculation function"""
        print("\n--- Grade Calculator ---")
        
        scores = []
        print("Enter subject scores (enter -1 to finish):")
        while True:
            try:
                score = float(input("Enter score (0-100): "))
                if score == -1:
                    break
                if score < 0 or score > 100:
                    print("Please enter a score between 0 and 100")
                    continue
                scores.append(score)
            except ValueError:
                print("Please enter a valid number")
        
        if scores:
            average = sum(scores) / len(scores)
            grade = self.calculate_grade(scores)
            print(f"\nAverage Score: {average:.2f}")
            print(f"Grade: {grade}")
        else:
            print("No scores entered.")
    
    def display_menu(self):
        """Display the main menu"""
        print("\n" + "=" * 40)
        print("      STUDENT RECORD SYSTEM")
        print("=" * 40)
        print("1. Add Student")
        print("2. View All Students")
        print("3. Search Student")
        print("4. Edit Student")
        print("5. Delete Student")
        print("6. Calculate Grade (Standalone)")
        print("7. Exit")
        print("-" * 40)
    
    def run(self):
        """Main program loop"""
        print("Welcome to the Student Record System!")
        
        while True:
            self.display_menu()
            
            choice = input("Enter your choice (1-7): ").strip()
            
            if choice == '1':
                self.add_student()
            elif choice == '2':
                self.view_all_students()
            elif choice == '3':
                self.search_student()
            elif choice == '4':
                self.edit_student()
            elif choice == '5':
                self.delete_student()
            elif choice == '6':
                self.calculate_grade_standalone()
            elif choice == '7':
                print("\nThank you for using the Student Record System. Goodbye!")
                break
            else:
                print("Invalid choice. Please enter a number between 1 and 7.")
            
            input("\nPress Enter to continue...")

# Run the program
if __name__ == "__main__":
    system = StudentRecordSystem()
    system.run()
```
