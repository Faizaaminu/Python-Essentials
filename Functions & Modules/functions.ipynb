Scope
--------------

>Scope is a variable that is only available from inside the region it is created.


__Local Scope__
>A variable created inside a function belongs to the local scope of that function, and can only be  used inside that function

```python
def my_func():
    x=300
    print(x)
my_func()
```

Function inside Function
+ the local variable can be access from a function within the function

```python
def my_func():
    x=300
    def myinnerfunc():
        print(x)
    myinnerfunc()

my_func()
```

__Global Scope__
>A variable created in the main body of the python code is a global variable and belongs to the global scope.</br>
global variable are available from within any scope,global and local

A variable created outside of a function is global and can used by anyone

```python
x=300
def my_func():
    print(x)

my_func()
print(x)
```

__Naming Variables__: operating with the same variable name inside and outside of a function, python will treat them as two seperate variables, one available in the global scope(outside the function) and one available in the local scope(inside the function)

```python
x=300
def my_func():
    x=200
    print(x)

my_func()
print(x)
```

__nonlocal Keyword__ 
Th ` nonlocal `keyword is used to work with  variables inside nested functions. it makes the variable belong to the outer function

```python

def my_func1():
    x="Jane"
    def my_func2():
        nonlocal x
        x="hello"
    my_func2()
    return x
print(my_func1())
```

Python Decorators
-----------------

A decorator is a function that takes another function as input and returns a new function


__Basic Decorators__: 
Define the decorator first, then apply it with `@decorator_name` above the function.

```python
"""" A basic decorator that uppercases the return value of the decorated function"""

def changecase(func):   #the function changecase is the decorator
    def inner():
        return func(). upper()
    return inner

@changecase
def my_func():                #the function my_func is the function that get decorated
    return "hello sally"
print(my_func())
```

__Multiple Decorator Calls__

```python
""""using the @changecase decorator on two functions"""

def changecase(func):
    def myinner():
        return func(). upper()
    return myinner

@changecase
def my_func():
    return "Hello Emil"

@changecase
def otherfunc():
    return "I am speedy !"
print(my_func())
print(otherfunc())
        
```

>Functions has metadata that can be access using the `__name__` and `__doc__` attribute

```python
""""using built in function `functools.wraps`  to preserve the original function,s name and docstring"""

import functools
def changecase(func):
    @functools.wraps(func)
    def myinner():
        return func().upper()
    return myinner

@changecase
def my_func():
    return "Have a great day! "

print(my_func.__name__)

```

Recursion
--------------

>Recursion is when a functio called itself.

__Base Case and Recursive Case__
- **base case**: A condition  that stops the recursion
- **recursive case**: A function calling itself with modified argument.

```python
"""" A simple recursion function that counts down from 5"""

def countdown(n):
    if n <= 0:
        print("Done!")
    else:
        print(n)
        countdown(n - 1)
countdown(5)
```

```python
def factorial(n):
    #Base case
    if n ==0 or n ==1:
        return 1
    #Recursive case
    else:
        return n * factorial(n - 1)
    
print(factorial(5))
```


__Fibonacci Sequence__
>The fibonacci sequence is a classic example where each number is the sum of the two preceding ones. the sequence start with 0 and 1:
0,1,1,2,3,5,8,13

```python
"""" Find the 7th number in the fibonacci sequence"""

def fibonacci(n):
    if n <=1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
print(fibonacci(7))
```

__Recursion with Lists__

```python
""""Calculate the sum of all elements in a list"""

def sum_list(numbers):
    if len(numbers) == 0:
        return 0
    else:
        return numbers[0] + sum_list(numbers[1:])
    
my_list=[1,2,3,4,5]
print(sum_list(my_list))
```

```python
""""Calculate the maximum value in a list"""

def find_max(numbers):
    if len(numbers) == 1:
        return 0
    else:
        max_of_rest=find_max(numbers[1:] )
        return numbers[0] if numbers[0] > max_of_rest else max_of_rest
    
my_list=[3,7,2,9,1]
print(find_max(my_list))
```

Generators
-----------

>Generators are functions that can pause and resume their execution.

```python


def my_gen():
    yield 1
    yield 2
    yield 3

for value in my_gen():
    print(value)
```

Instead using `return`, generators use the `yield` keyword.
`return` terminates the function, while `yield` pauses it and can be called multiple times


__The yield keyword__
>the `yield` keyword is what makes a function a generator

```python
def count_up_to(n):
    count=1
    while count <= n:
        yield count
        count += 1

for num in count_up_to(5):
    print(num)
```

```python
""""Using next() """

def my_gen():
    yield "Emil"
    yield "Tony"
    yield "Liwis"

gen=my_gen()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# When ther is no more value to yield , the generator raises a Stopiteration exception

def my_gen():
    yield 1
    yield 2
    

gen=my_gen()
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
# List comprehension - create a list

list_comp=[x * x for x in range(5)]
print(list_comp)

# Generator expression - create a generator
gen_exp=(x * x  for x in range(5))
print(gen_exp)
print(list(gen_exp))
```

__Fibonacci Sequence  Generator__

```python
"""Generate 100 Fibonacci numbers"""

def fibonacci():
    a, b =0, 1
    while True:
        yield a
        a, b=b, a + b

gen=fibonacci()
for _ in range(100):
    print(next(gen))
        
```

__Generator Method__
- `send()`Method: this method allows you to send a value to the generator
- `close()`  Method: this method stops the generator


```python
def echo_generator():
    while True:
        received= yield
        print(" Recieved:", received)

gen=echo_generator()
next(gen)  # Prime the generator
gen.send("Hello")
gen.send("World")
```

```python
def my_gen():
    try:
        yield 1
        yield 2
        yield 3
    finally:
        print(" Generator closed")

gen=my_gen()
print(next(gen))
gen.close()
```
