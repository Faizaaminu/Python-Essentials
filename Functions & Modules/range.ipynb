Python Range
-------------

> The built-in `range()` function returns an immutable sequence of numbers,commonly used for looping specific number of times

__Creating ranges__
The `range()` function can be with 1, 2 or 3 arguments.
range( start, stop, step)

Call range() with one Argument
+ If the range functio is called with only one argument, the argument represents the `stop` value

```python
# create a range of number from 0 to 9
x=range(10)
```

Call range() with two Argument
+ If the range function is called with two argument, the first argument represents the `start` value, and second argument represent the `stop` value

```python
# create a range of number from 0 to 9
x=range(3,9)
```

Call range() with three Argument
+ If the range function is called with three argument, the first argument represents the `start` value, and second argument represent the `stop` value then the third argument represent the `step` value

```python
# create a range of number from 0 to 9
x=range(3,9,2)
```

Ranges are often used in for loop to iterate a sequence of numbers

```python

for i in range(10):
    print(i)
```

Using List to display ranges

```python
print(list(range(5)))
print(list(range(1, 6)))
print(list(range(5, 20, 3)))
```

Slicing Ranges

```python
r=range(10)
print(r[2])
print(r[:3])
```

Membership Testing:using `in` operator to test if number are present in a range.
The return value is `True` when the number is present in the range and `False` when it is not.

```python
r=range(0,10,2)
print(6 in r)
print(7 in r)
```

```python
r=range(0,10,2)
print(len(r))     #Getting length of the range
```

__Array Method__
- `append()`: add an element at the end of the list.
- `clear()`: removes all the element from the list.
- `copy()`: return a copy of the list.
- `count()`: return the number of element with the specified value.
- `extend()`: add the element of a list(iterable) to the end of the current list.
- `index()`: return the index of the first element with specified value.
- `insert()`: adds an elementat the specified position.
- `pop()`: remove the element at the specified postion.
- `remove()`: remove the first item with the specified value.
- `reverse()`: reverse the order of the list
- `sort()`: sorts the list
